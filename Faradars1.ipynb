# ECDF
### Empirical Cumulative Distribution Function

```python
def ECDF(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
x, y = ECDF(smartphones.inch)
plt.figure(figsize=(11,8))
plt.scatter(x, y, s = 80)
plt.margins(0.05)
plt.xlabel('Inch', fontsize=20)
plt.ylabel('ECDF', fontsize=20)
plt.show()
```

## A Game

```python
step = 0
dice = np.random.randint(1,7)

if dice < 3:
    step = max(0, step-1)
elif dice <= 5:
    step = step+1
else:
    num = np.random.randint(1,7)
    step = step + num
    
print('dice is {} and you are in step {}'.format(dice, step))
        
```

## Random Walk

```python
step = 0
rand_walk = np.empty(0)
for i in range(100):
    dice = np.random.randint(1,7)

    if dice < 3:
        step = max(0, step-1)
    elif dice <= 5:
        step = step+1
    else:
        num = np.random.randint(1,7)
        step = step + num
    
    rand_walk = np.append(rand_walk, step)
    print('dice is {} and you are in step {}'.format(dice, step))
plt.step(np.arange(100), rand_walk)
plt.show()
```
